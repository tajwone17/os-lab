```python
n = int(input("Number of process in reference string : "))
print(n)
```

```python
rs = []
for i in range (n):
    x = int(input())
    rs.append(x)
    print(rs)
```

```python
m = int(input("number of frames : "))
print(m)
```

```python
fr = [-1 for i in range(m)]
print(fr)
```

```python
h= 0
ms = 0
idx = 0
for i in range(n):
    if rs[i] in fr:
        h += 1
    else:
        ms += 1
        fr[idx] = rs[i]
        idx += 1
        if(idx == m):
            idx = 0
```

```python
print(h)
print(ms)
```

```python
print("miss: ", ms/n*100, "%")
print("hit: ", h/n*100, "%")
```
