```python
r=int(input())
cols=7
matrix=[[0 for _ in range(cols)] for _ in range(r)]
for row in matrix:
    print(*row)
```

```python
for i in range (r):
  
    pid, at, bt = map(int, input().split())
    matrix[i][0]=pid
    matrix[i][1]=at
    matrix[i][2]=bt
for row in matrix:
    print(*row)

for i in range (r-1):
    for j in range(r-i-1): 
        if matrix[j][1]>matrix[j+1][1]:
            matrix[j],matrix[j+1]=matrix[j+1],matrix[j] 
time=0             
for i in range (r):
    at = matrix[i][1]
    bt = matrix[i][2]

    if time < at:  
        time = at  
    time+=bt         
    matrix[i][3]=time
    matrix[i][4]=time -at
    matrix[i][5]=matrix[i][4]-bt
    matrix[i][6]=matrix[i][5]
for row in matrix:
    print(*row)    
avg_ct=sum(row[3] for row in matrix) / r
avg_tat=sum(row[4] for row in matrix) / r     
print(f"avg copm time:{avg_ct:.2f}")
```
