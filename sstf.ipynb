```python
n = int(input("Number of requests: "))
requests = [int(input()) for _ in range(n)]
head = int(input("Enter initial head position: "))

seek_sequence = []
seek_count = 0
requests_copy = requests.copy()

current = head
while requests_copy:
    # Find the closest request
    distances = [abs(current - r) for r in requests_copy]
    min_idx = distances.index(min(distances))
    seek_count += distances[min_idx]
    current = requests_copy.pop(min_idx)
    seek_sequence.append(current)

print("Seek Sequence:", seek_sequence)
print("Total Seek Time:", seek_count)
```
