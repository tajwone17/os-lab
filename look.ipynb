```python
n = int(input("Number of requests: "))
requests = [int(input()) for _ in range(n)]
head = int(input("Enter initial head position: "))
direction = input("Enter direction (left/right): ").lower()

requests.append(head)
requests.sort()
seek_sequence = []
head_idx = requests.index(head)

if direction == "left":
    left = requests[:head_idx][::-1]
    right = requests[head_idx+1:]
    seek_sequence = left + right
else:
    right = requests[head_idx+1:]
    left = requests[:head_idx][::-1]
    seek_sequence = right + left

seek_count = 0
current = head
for track in seek_sequence:
    seek_count += abs(track - current)
    current = track

print("Seek Sequence:", seek_sequence)
print("Total Seek Time:", seek_count)
```
