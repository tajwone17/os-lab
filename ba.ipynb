```python
def banker_algorithm():
    # Get number of processes and resources
    n = int(input("Enter number of processes: "))
    r = int(input("Enter number of resources: "))
    
    # Input Allocation Matrix
    print("\nEnter Allocation Matrix (one process per line, space-separated):")
    alloc = []
    for i in range(n):
        row = list(map(int, input(f"P{i} allocation: ").split()))
        alloc.append(row)
    for row in alloc:
        print(*row)
    # Input MAX Matrix
    print("\nEnter MAX Matrix (one process per line, space-separated):")
    max = []
    for i in range(n):
        row = list(map(int, input(f"P{i} max: ").split()))
        max.append(row)
    
    # Input Available Resources
    print("\nEnter Available Resources (space-separated):")
    avail = list(map(int, input().split()))

    f = [0] * n
    ans = [0] * n
    ind = 0
    need = [[0] * r for _ in range(n)]

    for i in range(n):
        for j in range(r):
            need[i][j] = max[i][j] - alloc[i][j]

    y = 0
    for k in range(n):
        for i in range(n):
            if f[i] == 0:
                flag = 0
                for j in range(r):
                    if need[i][j] > avail[j]:
                        flag = 1
                        break

                if flag == 0:
                    ans[ind] = i
                    ind += 1
                    for y in range(r):
                        avail[y] += alloc[i][y]
                    f[i] = 1

    print("\nThe SAFE Sequence is as follows")
    for i in range(n - 1):
        print(f' P{ans[i]} ->', end='')
    print(f' P{ans[n - 1]}')

banker_algorithm()
```
