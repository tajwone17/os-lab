```python
cols = 7

n = int(input("Enter number of processes: "))

matrix = [[0 for _ in range(cols)] for _ in range(n)]

print("\nEnter 3 values for (Process No, Arrival Time, Burst Time):")
for i in range(n):
    print(f"{i+1}st process : ", end="")
    pid, at, bt = map(int, input().split())
    matrix[i][0] = pid
    matrix[i][1] = at
    matrix[i][2] = bt

for i in range(n-1):
    for j in range(n-i-1):
        if matrix[j][1] > matrix[j+1][1]:
            matrix[j], matrix[j+1] = matrix[j+1], matrix[j]


time = 0
for i in range(n):
    at = matrix[i][1]
    bt = matrix[i][2]

    if time < at:  
        time = at   

    time += bt
    matrix[i][3] = time                 
    matrix[i][4] = matrix[i][3] - at    
    matrix[i][5] = matrix[i][4] - bt    
    matrix[i][6] = matrix[i][5]         


print("\nMatrix Output (Process AT BT CT TT WT RT):")
for row in matrix:
    print(*row)


avg_CT = sum(row[3] for row in matrix) / n
avg_TT = sum(row[4] for row in matrix) / n
avg_WT = sum(row[5] for row in matrix) / n
avg_RT = sum(row[6] for row in matrix) / n

print("\nAverages:")
print(f"Average Completion Time : {avg_CT:.2f}")
print(f"Average Turnaround Time : {avg_TT:.2f}")
print(f"Average Waiting Time    : {avg_WT:.2f}")
print(f"Average Response Time   : {avg_RT:.2f}")
```
