```python
n = int(input("Number of process in reference string : "))
print(n)
```

```python
rs = []
for i in range(n):
    x = int(input())
    rs.append(x)
    print(rs)
```

```python
m = int(input("number of frames : "))
print(m)
```

```python
fr = [-1 for _ in range(m)]
print(fr)
```

```python
h = 0  
ms = 0 

for i in range(n):
    if rs[i] in fr:
        h += 1
    else:
        ms += 1
        if -1 in fr:
            fr[fr.index(-1)]=rs[i] 
        else:
            idx=-1
            arr = rs[i+1 :]
            flag=0
            for j in range(m):
                if fr[j] not in arr:
                    fr[j]=rs[i]
                    flag=1
                    break
                else:
                    if arr.index(fr[j]) > idx:
                        idx = arr.index(fr[j])
                        idx2 = j
            if flag==0:
                fr[idx2]=rs[i]

    print(fr)       
```

```python
print(h)
print(ms)
```

```python
print("miss: ", ms / n * 100, "%")
print("hit: ",  h / n * 100, "%")
```
