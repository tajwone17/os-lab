```python
n = int(input("Number of processes: "))
bt = []
for i in range(n):
    bt.append(int(input(f"Enter burst time for process {i+1}: ")))

wt = [0] * n
tat = [0] * n

# Sorting processes by burst time
processes = list(range(n))
processes.sort(key=lambda x: bt[x])

# Calculating TAT and WT
for i in range(n):
    if i == 0:
        tat[processes[i]] = bt[processes[i]]
    else:
        tat[processes[i]] = bt[processes[i]] + tat[processes[i-1]]
    wt[processes[i]] = tat[processes[i]] - bt[processes[i]]

print("\nProcess\tBurst Time\tWaiting Time\tTurnaround Time")
for i in range(n):
    print(f"P{i+1}\t{bt[i]}\t\t{wt[i]}\t\t{tat[i]}")

print(f"\nAverage Waiting Time: {sum(wt)/n}")
print(f"Average Turnaround Time: {sum(tat)/n}")
```
